```python
# Function for join table
def readMyStream_test_quiz2(rdd, rdd2):
  if not rdd.isEmpty() and not rdd2.isEmpty():
    df = spark.read.json(rdd)
    df2 = spark.read.json(rdd2)

    # The inferred schema can be visualized using the printSchema() method
    df.printSchema()
    df2.printSchema()

    # Creates a temporary view using the DataFrame
    df.createOrReplaceTempView("kvs5_")
    df2.createOrReplaceTempView("kvs6_")

    # SQL statements can be run by using the sql methods provided by spark
    #resultDF = spark.sql("SELECT kvs3_.id, kvs4_.firstname ,kvs3_travel FROM kvs3_ LEFT JOIN kvs4_ ON kvs3_.id = kvs4_.id ")
    resultDF = spark.sql("SELECT id, SUM(num_click) as num_click from kvs5_ GROUP BY id")
    resultDF.show()

    resultDF2 = spark.sql("SELECT * from kvs6_")
    resultDF2.show()

    resultDF3 = spark.sql("SELECT Country, SUM(num_click) AS Total_click FROM (SELECT kvs5_.id, kvs5_.num_click, kvs6_.Country from kvs5_ LEFT JOIN kvs6_ ON kvs5_.id = kvs6_.id ORDER BY kvs5_.id) GROUP BY Country ")
    resultDF3.show()
```

```python
# config topic
kafka_bootstrap_servers = 'ec2-13-229-46-113.ap-southeast-1.compute.amazonaws.com:9092'

kafka_topic_name = "topic-dads6005-4-1"
kafka_topic_name2 = "topic-dads6005-4-2"
kafka_topic_name3 = "topic-dads6005-4-3"

kafka_topic_name_ = "topic-dads6005-4-4"

kafka_topic_name_cin_ = "cin3"

kafka_topic_name_aof = "aof"

kafka_topic_name_aof_2 =  "aof23"    #"aof12"
kafka_topic_name_aof_3= "aofSource23" #"aofSource11"
```

```python
# start pyspark
sc = pyspark.SparkContext()
spark = SparkSession(sc)
ssc = StreamingContext(sc,5) # Show results every 5 seconds
```

```python
# main
kvs5 = KafkaUtils.createDirectStream(ssc, 
                                    [kafka_topic_name_aof_2], # kafka_topic_name3
                                    {
                        'bootstrap.servers':kafka_bootstrap_servers,
                        'group.id':'test-group',
                        'auto.offset.reset':'largest'
                        })

kvs6 = KafkaUtils.createDirectStream(ssc, 
                                    [kafka_topic_name_aof_3],   
                                    {
                        'bootstrap.servers':kafka_bootstrap_servers,
                        'group.id':'test-group',
                        'auto.offset.reset':'largest'
                        })

kvs5.pprint(50)
kvs6.pprint(50)

kvs5_ = kvs5.map(lambda x: x[1]).window(10, 5)
kvs6_ = kvs6.map(lambda x: x[1]).countByValue().map(lambda x: x[0]).window(10, 5)

kvs5_.pprint(50)
kvs6_.pprint(50)

joinDstream = kvs5_.transformWith(readMyStream_test_quiz2, kvs6_)
joinDstream.pprint(100)


ssc.start()
ssc.awaitTerminationOrTimeout(60)
ssc.stop()
sc.stop()
```
